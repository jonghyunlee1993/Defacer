```python
import json
import os
import glob
```

```python
dir_root = 'D:\SNS_raw'
os.chdir(dir_root)
sub_list = glob.glob("sns*")

final = []

for sub in sub_list:
    dir_sub = os.path.join(dir_root, sub)
    os.chdir(dir_sub)
    try:
        comment_fname = 'comments_' + sub + '.json'
        like_fname = 'likes_' + sub + '.json'
        connection_fname = 'connections_' + sub + '.json'

        with open(comment_fname, encoding='UTF-8') as comment_json:
            comment_data = json.load(comment_json)

        with open(like_fname, encoding='UTF-8') as like_json:
            like_data = json.load(like_json)

        with open(connection_fname, encoding='UTF-8') as connection_json:
            connection_data = json.load(connection_json)

        num_comment = len(comment_data["media_comments"])
        num_like = len(like_data["media_likes"])
        num_follower = len(connection_data["followers"])
        num_following = len(connection_data["following"])

        result = [sub, num_comment, num_like, num_follower, num_following]
        

        if os.path.isfile(os.path.join(dir_sub, 'comments_' + sub + '_2.json')):
            with open(comment_fname, encoding='UTF-8') as comment_json:
                comment_data_2 = json.load(comment_json)

            with open(like_fname, encoding='UTF-8') as like_json:
                like_data_2 = json.load(like_json)

            with open(connection_fname, encoding='UTF-8') as connection_json:
                connection_data_2 = json.load(connection_json)

            num_comment_2 = len(comment_data_2["media_comments"])
            num_like_2 = len(like_data_2["media_likes"])
            num_follower_2 = len(connection_data_2["followers"])
            num_following_2 = len(connection_data_2["following"])

            result = [sub, num_comment + num_comment_2, num_like + num_like_2, 
                      num_follower + num_follower_2, num_following + num_following_2]
            
#             final.append(result_2)

            if os.path.isfile(os.path.join(dir_sub, 'comments_' + sub + '_3.json')):
                with open(comment_fname, encoding='UTF-8') as comment_json:
                    comment_data_3 = json.load(comment_json)

                with open(like_fname, encoding='UTF-8') as like_json:
                    like_data_3 = json.load(like_json)

                with open(connection_fname, encoding='UTF-8') as connection_json:
                    connection_data_3 = json.load(connection_json)

                num_comment_3 = len(comment_data_3["media_comments"])
                num_like_3 = len(like_data_3["media_likes"])
                num_follower_3 = len(connection_data_3["followers"])
                num_following_3 = len(connection_data_3["following"])

                result = [sub,  num_comment + num_comment_2 + num_comment_3, num_like + num_like_2 + num_like_3, 
                            num_follower + num_follower_2 + num_follower_3, num_following + num_following_2 + num_following_3]
#                 final.append(result_3)

        final.append(result)
    
    except:
        print("Unexpected Error Occured")
        
```

```python
import pandas as pd

my_result = pd.DataFrame(final, columns = ['Code', 'Num_comment', 'Num_like', 'Num_follower', 'Num_following'])
print(my_result)

my_result.to_csv(os.path.join(dir_root ,'meta_data_result.csv'), index = True)
```
