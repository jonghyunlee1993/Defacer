```python
import numpy as np
import pandas as pd
from collections import Counter

data_path = "../data/upload_time.csv"
```

```python
data = pd.read_csv(data_path)
data.head()
```

```python
sub_list = data['Code'].unique().tolist()
print(sub_list)
```

```python
result = pd.DataFrame(columns = range(24))

for sub in sub_list:
    sub_res = Counter(data[data['Code'] == sub]['Hour'])
    
    key = list(sub_res.keys())
    val = list(sub_res.values())
    
    temp = pd.DataFrame(val).T
    temp.columns = key
    result = result.append(temp)
    
# #     items = sub_res.items()
# #     final = [item[1] for item in sorted(items, key = lambda x: x[0])]
```

```python
result["Code"] = sub_list
result = result.fillna(0)
result.to_csv("../data/upload_time_result_by_sub.csv")
```

```python
# 3 시간 단위로
result = pd.DataFrame(columns = range(8))

for sub in sub_list:
    sub_counter = Counter(data[data['Code'] == sub]['Hour'])
    
    sub_key = list(sub_counter.keys())
    sub_value = list(sub_counter.values())
    
    print(sub_key)
    print(sub_value)
    
    sub_res = [0, 0, 0, 0, 0, 0, 0, 0]
    
    for hour in range(24):
        try:
            if hour in [0, 1, 2]:
                sub_res[0] += sub_value[sub_key.index(hour)]
            elif hour in [3, 4, 5]:
                sub_res[1] += sub_value[sub_key.index(hour)]
            elif hour in [6, 7, 8]:
                sub_res[2] += sub_value[sub_key.index(hour)]
            elif hour in [9, 10, 11]:
                sub_res[3] += sub_value[sub_key.index(hour)]
            elif hour in [12, 13, 14]:
                sub_res[4] += sub_value[sub_key.index(hour)]
            elif hour in [15, 16, 17]:
                sub_res[5] += sub_value[sub_key.index(hour)]
            elif hour in [18, 19, 20]:
                sub_res[6] += sub_value[sub_key.index(hour)]
            elif hour in [21, 22, 23]:
                sub_res[7] += sub_value[sub_key.index(hour)]
        except:
            print(f"{sub}'s {hour} upload data not in list!")
            
    temp = pd.DataFrame(sub_res).T
    result = result.append(temp)
```

```python
print(result)
```

```python
result.to_csv('../data/upload_time_result_three_hour_binding.csv')
```
