```python
import os 
import csv
import glob

os.chdir('H:/SNS_Data/')
dir_root = os.getcwd()
dir_save = ('H:/')
Subj = os.listdir()

for sub in Subj:
    os.chdir(sub)
    Images = glob.glob('*.jpg')
    os.chdir(sub + '_cropped_face')
    Faces = glob.glob('Face*.jpg')
    ImageNum = len(Images)
    FaceNum = len(Faces)
    
    print(sub + ' has ' + str(ImageNum) + ' images and ' +  str(FaceNum) + ' face images')
    Results = [sub, ImageNum, FaceNum]
    
    os.chdir(dir_root)
    
    with open(dir_save + 'Face_vs_Images.csv', 'a', newline='') as f:
        writer = csv.writer(f)
        writer.writerow(Results)
```
