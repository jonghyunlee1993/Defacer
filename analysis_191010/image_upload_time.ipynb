```python
import json
import os
import time
import glob
import datetime
import pandas as pd
import matplotlib.pyplot as plt

def getDayName(year, month, day):
    dayString = ["Mon", "Tue", "Wen", "Thu", "Fri", "Sat", "Sun"]
    
    year = int(year)
    month = int(month)
    day = int(day)
    
    return dayString[datetime.date(year, month, day).weekday()]
```

```python
dir_root = r'D:\SNS_raw'
os.chdir(dir_root)
sub_list = glob.glob('sns*')

final = []

for sub in sub_list:
    dir_sub = os.path.join(dir_root, sub)
    os.chdir(dir_sub)
    
    try:
        media_fname = 'media_' + sub + '.json'
        
        with open(media_fname, encoding='UTF-8') as media_json:
            media_data = json.load(media_json)
            
            for ind, post in enumerate(media_data['photos']):
                upload_time = post['taken_at']
                
                year = upload_time[0:4]
                month = upload_time[5:7]
                day = upload_time[8:10]
                hour = upload_time[11:13]
                minute = upload_time[14:16]
                
                DOW = getDayName(year, month, day)
                
                post_result = [sub, ind, DOW, hour, minute]
                final.append(post_result)
    except:
        print('Unexpected Error occured')

        
final = pd.DataFrame(final, columns = ['Code', 'Post idx', 'DOW', 'Hour', 'Min'])
print(final)
```

```python
final.to_csv(os.path.join(dir_root, 'upload_time.csv'))
```
